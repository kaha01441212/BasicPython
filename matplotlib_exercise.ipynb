## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df = pd.read_csv('data/scatter_data.csv')

df_class_0 = df[df['class'] == 0]
df_class_1 = df[df['class'] == 1]
df_class_2 = df[df['class'] == 2]

plt.scatter(df_class_0['length (cm)'], df_class_0['width (cm)'] ,color = 'red' ,label='Class0')
plt.scatter(df_class_1['length (cm)'], df_class_1['width (cm)'] ,color = 'blue' ,label='Class1')
plt.scatter(df_class_2['length (cm)'], df_class_2['width (cm)'] ,color = 'green' ,label='Class2')

plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('クラス0〜2のlengthとwidchに関する散布図')

plt.grid()
plt.legend(['Class0', 'Class1','Class2'])
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df = pd.read_csv('data/line_data.csv')

plt.figure(figsize=(6, 10))

plt.subplot(2, 1, 1)
plt.plot(df['期間'], df['件数'], marker="o")
plt.ylabel('件数')
plt.xlabel('期間')
plt.title('期間ごとの件数の推移')
plt.xticks(rotation=30)
plt.grid()


plt.subplot(2, 1, 2)
plt.bar(df['期間'], df['費用'])
plt.ylabel('費用(円)')
plt.xlabel('期間')
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=30)
plt.grid()

plt.subplots_adjust(hspace=0.5)
plt.show()
```
